### **a.**

*Extraer la información del archivo.*

```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = 6

vaccinations = pd.read_csv("country_vaccinations.csv")
vaccinations
```

<hr>

### **b.**

*Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones
puedes realizar con cada una de ellas, asegurandote que **las columnas con fechas sean del
tipo datetime64**.*

```python
vaccinations["date"] = pd.to_datetime(vaccinations["date"])
vaccinations.info()
```

<hr>

### **c.**

*Determinar la **cantidad de vacunas aplicadas de cada compañía** (en base a como lo reporta cada país en la columna **vaccines**).*

```python
vaccinations[["total_vaccinations", "vaccines"]].groupby("vaccines").sum()
```

<hr>

### **d.**

*Obtener la **cantidad de vacunas aplicadas en todo el mundo**.*

```python
aplicadas_mundo = vaccinations["total_vaccinations"].sum().astype("int64")
print(f"Cantidad de vacunas aplicadas en todo el mundo: {aplicadas_mundo}")
```

<hr>

### **e.**

*Calcular el **promedio de vacunas aplicadas por país**.*

```python
promedio_aplicadas_pais = int(vaccinations[["iso_code", "total_vaccinations"]].groupby("iso_code").sum().mean())
print(f"Promedio de vacunas aplicadas por país: {promedio_aplicadas_pais}")
```

<hr>

### **f.**

*Determinar la **cantidad de vacunas aplicadas el día 29/01/21** en todo el mundo.*

```python
fecha = "2021-01-29"
aplicadas_fecha = int(vaccinations[vaccinations["date"] == fecha]["total_vaccinations"].sum())
print(f"Cantidad de vacunas aplicadas el {fecha}: {aplicadas_fecha}")
```

<hr>

### **g.**

*Crear un dataframe nuevo denoninado **"conDiferencias"** que contenga los datos originales y una columna derivada (**"diferencias"**) con las diferencias de aplicación entre las columnas
**daily_vaccionations** y **daily_vaccionations_raw**.*

```python
conDiferencia = vaccinations.copy()
conDiferencia["diferencias"] = conDiferencia["daily_vaccinations"] - conDiferencia["daily_vaccinations_raw"]
conDiferencia
```

<hr>

### **h.**

*Obtener **el período de tiempo entre** el registro con fecha más reciente y el registro con
fecha más antigua.*

```python
periodo_mayor = vaccinations["date"].max() - vaccinations["date"].min()
print(f"Periodo de tiempo mayor entre registros: {periodo_mayor.days} días")
```

<hr>

### **i.**

*Crear un dataframe nuevo denominado **"conCantidad"** que contenga los datos originales y una columna derivada (**"canVac"**) **con la cantidad de vacunas utilizadas cada día** (usar la columna "vaccines" y separar por el carácter / ).*

```python
conCantidad = vaccinations.copy()  ### Al copiar el original, se restablece el tipo de dato para "date", por lo que si no funciona hay que correr el punto "b" nuevamente y volver a intentar
conCantidad["canVac"] = conCantidad["vaccines"].str.split("/").str.len()
conCantidad
```

<hr>

### **j.**

*Generar un dataframe denominado **"antes20"** con todos los registros que se hayan realizado **antes del 20 de diciembre de 2020**.*

```python
antes20 = vaccinations[vaccinations["date"] < "2020-12-20"]
antes20
```

<hr>

### **k.**

*Obtener un dataframe denominado **"pfizer"** con todos los registros donde se haya utilizado la vacuna Pfizer.*

```python
pfizer = vaccinations[vaccinations["vaccines"].str.contains("Pfizer")]
pfizer
```

<hr>

### **l.**

*Almacenar los dataframes generados (**"conDiferencias", "conCantidad", "antes20" y "pfizer"**) en un archivo de Excel denominado **"resultadosReto.xlsx"**, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.*

```python
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
  conDiferencia.to_excel(writer, sheet_name="conDiferencia")
  conCantidad.to_excel(writer, sheet_name="conCantidad")
  antes20.to_excel(writer, sheet_name="antes20")
  pfizer.to_excel(writer, sheet_name="pfizer")
```
